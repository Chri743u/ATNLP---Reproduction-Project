```python
def load_scan(path):
    pairs = []
    with open(path, "r") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            # Example:
            # IN: jump around left twice OUT: LTURN JUMP LTURN JUMP
            inp, out = line.split(" OUT: ")
            inp = inp.replace("IN: ", "")
            pairs.append((inp, out))
    return pairs

train_data = load_scan(r"C:\Users\gabri\OneDrive\Skrivebord\SCAN\simple_split\tasks_train_simple.txt")
test_data  = load_scan(r"C:\Users\gabri\OneDrive\Skrivebord\SCAN\simple_split\tasks_test_simple.txt")
```
